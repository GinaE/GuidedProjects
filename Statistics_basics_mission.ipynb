Make a line plot with day_numbers on the x axis and cars_in_parking_lot on the y axis.

```python
day_numbers = [1,2,3,4,5,6,7]
snail_crawl_length = [.5,2,5,10,1,.25,4]
cars_in_parking_lot = [5,6,4,2,1,7,8]

import matplotlib.pyplot as plt

%matplotlib inline 

fig, ax = plt.subplots()
line1 = ax.plot(day_numbers, snail_crawl_length, color='b')
line2 = ax.plot(day_numbers, cars_in_parking_lot, color='g')
ax.set_title("daily measurements")
ax.legend( (line1[0], line2[0]), ('snail crawl length','# of cars in lot') )
plt.show()
```
